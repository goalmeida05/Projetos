# Seleção de Frequências
a seguradora deseja que realizemos algumas seleções e retire a frequência de cada uma delas. Recebemos o seguinte e-mail da empresa:



*   Selecione somente os imóveis classificados com tipo 'Apartamento'.
*   Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
*   Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
*   Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

(Obs: Utilizaremos o arquivo que geramos no Relatório de análise III)



```python
import pandas as pd
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
#dados.head(10) #verificação
```

# Selecione somente os imóveis classificados com tipo 'Apartamento'.



```python
selecao = dados['Tipo'] == 'Apartamento'
#selecao #Conferindo se a seleção foi feita corretamente
n1 = dados[selecao].shape[0] #Com n1 saberemos quantos apartamentos estão na lista
n1
```

# Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
Agora temos mais de um tipo de coleção para analisar, casa, casa de condomínio e casa de vila. Usaremos o operador | que significa "ou".

```python
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao].shape[0]
n2
```

# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

```python
selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n3 = dados[selecao].shape[0]
n3
```

# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

```python
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000.0)
n4 = dados[selecao].shape[0]
n4
```

# Relatório Final

```python
print(f"Imóveis classificados apartamentos: {n1}")
print(f"Imóveis como Casa, Casa de condomínio e Casa de vila são: {n2}")
print(f"Imóveis que possuem área entre 60 a 100 metros: {n3}")
print(f"Imóveis que tenham pelo menos 4 quartos e aluguel menor que R$2000,00: {n4}")
```

# Crie um DataFrame somente com os alunos aprovados.

```python
selecao = alunos['Aprovado'] == True
aprovados = alunos[selecao]
aprovados
```

```python
selecao = (alunos['Sexo'] == 'F') & (alunos['Aprovado'] == True)
alunas_aprovadas = alunos[selecao]
alunas_aprovadas
```
