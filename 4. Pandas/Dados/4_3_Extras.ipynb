```python
import pandas as pd
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
#list('321') #Relembrando que esse comando irá criar uma lista com os números 3,2,1 em forma de string.
df = pd.DataFrame(data, list('321'), list('ZYX'))
df #As listas armazenadas na lista data tem o list('321') como index e o list('ZYX') nomeando cada coluna
```

```python
df.sort_index(inplace = True)
df #Aqui organizamos o index em ordem crescente
```

```python
df.sort_index(inplace = True, axis = 1)
df #Organizamos também as colunas em ordem alfabética
```

```python
df.sort_values(by = 'X', inplace = True)
df #Organizando a tabela colocando os valores da coluna X em ordem crescente
```

```python
df.sort_values(by = ['X','Y'], inplace = True)
df #Organizando a tabela colocando os valores da coluna X e Y em ordem crescente
```

```python
df.sort_values(by = '3', axis = 1, inplace = True)
df #Organizando os valores do index 3 em ordem crescente.
```
