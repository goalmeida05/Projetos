Tipos de imóveis

```python
import pandas as pd
dados = pd.read_csv('aluguel.csv', sep = ';')
tipo_de_imovel = dados['Tipo'] #Como vamos trabalhar somente com os tipos de imóveis separamos ela em uma nova variável e isso facilitará nosso trabalho
Signature: tipo_de_imovel.drop_duplicates(keep='first', inplace=False) #fazendo isso estamos agrupando os iguais e então temos uma nova tabela com o imovel e quantidade desses presente no arquivo.
```

Quando temos o keep configurado como first, todas as duplicatas são eliminadas. Temos, ainda, o inplace. Quando seu dafault esta marcado como true, ele modifica a variável a partir da execução do método, não precisamos criar uma variável e atribuir a execução à ela.
