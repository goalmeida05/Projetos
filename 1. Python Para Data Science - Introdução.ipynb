Vamos trabalhar agora com listas e como acessar suas informações.

Uma lista é capaz de armazenar vários elementos, podemos adcionar mais, excluir e acessar ela completa ou parte dela.

```python
idades = [10,20,30,18,50] #definimos uma lista com esses elementos contídos
print(idades[0:2]) #irá imprimir os elementos 0 e 1 
# E por quê não imprime 10, 20 e 30, já que deveria ser os elementos da posição 0 até 2?
#Isso ocorre porque o último número requisitado não está contido no conjunto de 0 até 2 
#Com a segunda posição fora do conjunto.

print(idades[1:]) #Pegaremos todas as idades com excessão da primeira.
print(idades[-3]) #Somente a última.

```

```python
def verifica_se_pode_dirigir(idade):
  if idade >= 18:
    print(f'{idade} anos de idade, TEM permissão para dirigir')
  else:
    print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')

for idade in idades: #para cara elemento idade, na lista e idades é chamado a função de comparação 
  verifica_se_pode_dirigir(idade)
```
