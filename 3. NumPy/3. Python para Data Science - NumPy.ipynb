```python
import numpy as np #Importanto a biblioteca NumPy e colocando um apelido nela que nos poupará tempo

km = np.loadtxt("carros-km.txt") #O loadtxt carrega o arquivo TXT passado por parâmetro dentro de um array numpy
#km #Caso queira testar se o parâmetro foi passado corretamente tire o comentário anterior que antecede o km e rode o código.

anos = np.loadtxt('carros-anos.txt', dtype = int) #Definindo que os anos devem ser números inteiros e não pontos flutuantes
```

"Agora que conhecemos um pouco do funcionamento desses arrays, ainda falta entendermos o motivo para utilizá-los ao invés das listas.
```python
np_array = np.arange(1000000) #Um array preenchido com os números de 0 até 1000000 -1
py_list = list(range(1000000)) #Uma lista preenchida com os números de 0 até 1000000 -1

#vamos comparar o tempo de execução desses códigos

%time for _ in range(100): np_array *=2 #%time = verifica o tempo de execução desse for. 
# _ ignora a variável que estaria ali 
# todo esse for irá multiplicar diretamente os valores do array por 2.
```

```python
#Repetiremos a ideia acima, mas para a lista
%time for _ in range(100): py_list = [x*2 for x in py_list] #Irá multiplicar cada elemento dessa lista por 2.
```

Agora veremos alguns métodos que podemos aplicar para acrescentar ou retirar termos das listas e como oganizá-las. 
Aqui trabalharemos com os métodos sort(), append(), pop() e copy().


```python
Acessorios = ['Rodas de liga', 'Travas elétricas', 'Piloto automático', 'Bancos de couro', 'Ar condicionado', 'Sensor de estacionamento', 'Sensor crepuscular', 'Sensor de chuva']
```

```python
Acessorios.sort() #Esse método irá reorganizar a lista acessórios em ordem alfabética.
Acessorios
```

```python
Acessorios.append('4 x 4') #Estamos acrescentando o elemento 4 x 4 na list acessórios.
```

```python
Acessorios.pop() #Caso não adicione nenhum parâmetro dentro do pop ele irá eliminar o último elemento da lista.
```

```python
acessorios_2 = Acessorios.copy()
#Nesse caso não podemos fazer simplesmente que acessorios_2 = Acessorios
#Pois qualquer coisa que fizemos numa lista irá alterar a outra
#Logo se faz necessário o uso do .copy()
```

```python
#Testes com listas
dados = [ 
    ['A', 'B', 'C'],
    ['D', 'E', 'F'],
    ['G', 'H', 'I']
]
```

```python
result = []
for lista in dados:
    for item in lista:
        result.append(item)
result
```
